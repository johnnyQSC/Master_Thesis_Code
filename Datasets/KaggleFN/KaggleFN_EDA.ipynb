```python
#載入資料集
import pandas as pd
file=pd.read_csv('KaggleFN.csv')
file
```

```python
#找出缺漏值
file.isna().sum()
```

```python
#刪除具有缺漏值的欄位
file_dropna=file.dropna()
file_dropna
```

```python
#觀察type欄位的資料分佈
file_dropna['type'].hist()
print(file_dropna['type'].value_counts())
```

```python
#觀察language欄位的資料分佈
file_dropna['language'].hist()
print(file_dropna['language'].value_counts())
```

```python
#觀察hasImage欄位的資料分佈
file_dropna['hasImage'].hist()
print(file_dropna['hasImage'].value_counts())
```

```python
#觀察label欄位的資料分佈
file_dropna['label'].hist()
print(file_dropna['label'].value_counts())
```

```python
#將type與language欄位資料型態轉換為數字
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
file_dropna['type']=labelencoder.fit_transform(file_dropna['type'])
file_dropna['language']=labelencoder.fit_transform(file_dropna['language'])
```

```python
#type敘述統計分析與盒鬚圖觀察離群值
print("type敘述統計")
print(file_dropna['type'].describe())
import matplotlib.pyplot as plt
plt.boxplot(file_dropna['type'],showmeans=True)
plt.title('type')
plt.show()
```

```python
#language敘述統計分析與盒鬚圖觀察離群值
print("language敘述統計")
print(file_dropna['language'].describe())
plt.boxplot(file_dropna['language'],showmeans=True)
plt.title('language')
plt.show()
```

```python
file_drop=file_dropna.drop(['author','published','title','text','language','site_url','main_img_url','type','hasImage'],axis=1)
file_drop
```

```python
#另存資料集
file_drop.to_csv("KaggleFN_EDA.csv",index=False,encoding='utf-8')
```
